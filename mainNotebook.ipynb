## 1.3.

On remarque que le reste par la division euclidienne de $code$ donne le code num√©rique du premier caract√®re.

Posons $code = \displaystyle\sum_{i=0}^{n-1} c(b_i) L^i$ o√π $c(b_i) \in [0; L-1]$

$code = c(b_0) L^0 + c(b_1) L^1 + ... + c(b_{n-1}) L^{n-1}$

$code = c(b_0) + c(b_1) L^1 + ... + c(b_{n-1}) L^{n-1}$

Ainsi le reste de la division euclidienne de $code$ par L est : $c(b_0)$

Posons $code' = \frac{(code - c(b_0))}{L}$

$code' = \displaystyle\sum_{i=1}^{n-1} \frac{c(b_i) L^i}{L}$

$code' = \displaystyle\sum_{i=1}^{n-1} c(b_i) L^{i-1}$

$code' = \displaystyle\sum_{i=0}^{n-2} c(b_i) L^{i}$

On r√©it√®re alors le proc√©d√©, on remarque que ce syst√®me de codage poss√®de un effet de bord, si le caract√®re de code num√©rique est pr√©sent (plusieurs fois) √† la fin du bloc, celui-ci ne sera pas retranscrit.

```python
def DecodageCodeBloc(code):
    bloc = ""

    if code < 0:
        raise ValueError(f"code: {code} < 0 !")
    elif code == 0:
        bloc = "\n"
    else:
        i = 0
        while code > 0:
            codeChar = code % L
            char = dicoDecodage[codeChar]
            bloc = bloc + char
            code = (code - codeChar) // L
            i = i + 1

    return bloc
```

```python
bloc = "Bonjour\nComment allez-vous ?\nCordialement"
code = EncodageBloc(bloc)
print(f"Le bloc original est:\n{bloc}.\nSon code associ√© est {code}.")

blocDecode = DecodageCodeBloc(code)
print(
    f"Le d√©codage du code associ√© donne:\n{blocDecode}.\nLes 2 blocs sont{'' if bloc == blocDecode else ' non'} identiques."
)
```

```python
def Encodage(message, l):
    codes = []
    for pas in range(0, len(message), l):
        bloc = message[pas : pas + l]
        code = EncodageBloc(bloc)
        codes.append(code)
    return codes


def Decodage(codes):
    message = ""
    for code in codes:
        bloc = DecodageCodeBloc(code)
        message = message + bloc
    return message


message = "Epigraphe du Gai Savoir\nJe vis dans ma propre maison\nDe personne jamais ne fus contrefa√ßon\nEt de tout ma√Ætre j'ai bien ri\nQui ne sait pas rire de lui.\n(Inscrit sur ma porte d'entr√©e)"

l = 9
print(f"Le message original est :\n{message}")
messageEncode = Encodage(message, l)
print(f"Le message encod√© par segment de {l} est :\n{messageEncode}")
messageDecode = Decodage(messageEncode)
print(
    f'Le d√©codage du code associ√© est:\n{blocDecode}.\nLes 2 blocs sont{"" if message == messageDecode else " non"} identiques.'
)
```

## 2.1. Algorithme d'Euclide en version r√©cursive

Page 23

```python
def euclideRec(a, b):
    # Retourne le PGCD de a et b

    if b == 0:
        return a
    else:
        return euclideRec(b, a % b)


euclideRec(17, 10)
```

## 2.2. Algorithme d'Euclide √©tendu en version it√©rative
Page 25

```python
def EuclideEtendu(a, b):
    # Retourne le PGCD de a et b, et les coefficients de B√©zout

    r0 = a
    u0 = 1
    v0 = 0

    r1 = b
    u1 = 0
    v1 = 1

    while r1 != 0:
        q = r0 // r1

        (r0, u0, v0, r1, u1, v1) = (r1, u1, v1, r0 - q * r1, u0 - q * u1, v0 - q * v1)

    return r0, u0, v0


print(EuclideEtendu(5, 17))
```

## 2.3. Inversion modulo

Page 30

```python
def InverseModulo(x, n):
    # Retourne un entier y tel que x * y est congru √† 1 mod n
    r0 = n
    v0 = 0

    r1 = x
    v1 = 1

    while r1 != 0:
        q = r0 // r1
        (r0, v0, r1, v1) = (r1, v1, r0 - q * r1, v0 - q * v1)

    if r0 != 1:
        raise ValueError(f"{x} n'est pas inversible modulo {n}")
    else:
        return v0


InverseModulo(3, 11)
```

## 2.4. Th√©or√®me des restes chinois
Page 33

```python
def RestesChinois(a, m):
    # a : reste
    # m : longueur des rangs

    p = 1
    for n in m:
        p = p * n
    s = 0

    for i in range(len(a)):
        mIp = p // m[i]
        mIpInv = InverseModulo(mIp, m[i])
        s = s + a[i] * mIp * mIpInv

    return s % p


print(RestesChinois([3, 1, 2, 6, 0], [5, 8, 9, 11, 13]))
```

## 2.5. Exponentiation modulaire rapide
Page 54

```python
def ExpoModRapide(x, y, n):
    # x^y mod n
    res = 1

    while y > 0:
        if y % 2 == 0:
            x = (x * x) % n
            y = y // 2
        else:
            res = (x * res) % n
            y = y - 1

    return res


ExpoModRapide(4, 13, 497)
```

## 3.1 G√©n√©ration de nombres premier

```python
def TestMiller_base(n: int, a: int) -> bool:
    """
    Effectue le test de Miller-Rabin sur l'entier n en base a.

    Args:
        n (int): L'entier √† tester
        a (int): La base

    Returns:
        bool: n est-il premier en base a ?
    """
    r = 0
    s = n - 1

    while s % 2 == 0:  # En sortie de la boucle : n = 2^r * s + 1 et s impair
        r = r + 1
        s = s // 2

    aPuissance = ExpoModRapide(a, s, n)

    # Condition 1 et
    if aPuissance == 1:
        return True

    # Condition 2
    if aPuissance == n - 1:  # Condition 2 pour i = 0
        return True

    for i in range(1, r):  # i varie de 1 √† r-1
        aPuissance = ExpoModRapide(aPuissance, 2, n)
        if aPuissance == n - 1:  # Condition 2 pour i
            return True

    return False


print(TestMiller_base(221, 174))
print(TestMiller_base(221, 137))
print(TestMiller_base(5, 4))
```

```python
def TestMiller(n: int, m: int) -> bool:
    """
    Effectue le test de Miller-Rabin sur l'entier n avec m bases choisies al√©atoirement entre 0 et n-1

    Args:
        n (int): L'entier √† tester
        m (int): Le nombre de bases √† choisir al√©atoire

    Returns:
        bool: n a-t-il pass√© le test de Miller-Rabin m fois ?
    """

    for i in range(m):
        a = random.randint(1, n - 1)
        if not (TestMiller_base(n, a)):
            return False

    return True
```

Un nombre cod√© sur $k$ bits $n$ v√©rifie : $n \in [ 2^{k-1}; 2^{k} [ = [ 2^{k-1}; 2^{k}-1 ]$.

```python
def Generateur_Premier(k: int, epsilon: int) -> int:
    """
    Retourne un nombre pseudo-premier cod√© sur k bits

    Args:
        k (int): Le nombre de bits sur lequel cod√© le nombre pseudo-premier
        epsilon (float): La probabilit√© seuile

    Returns:
        int: Un entier qui √† une probabilit√© epsilon de ne pas √™tre entier.
    """

    n = random.randint(
        2 ** (k - 1),
        2**k,
    )

    m = ceil(log((k * log(2)) / (2 * epsilon), 4))

    while not (TestMiller(n, m)):
        n = random.randint(
            2 ** (k - 1),
            2**k,
        )

    return n
```

### 4.1.1 Construction de *p*

#### 4.1.1.1.
Le chiffrement d'ElGamal impose $p$ un nombre premier pour lequel l'inversion de l'exponentielle modulaire est difficile. C'est-√†-dire :
- $p$ est un tr√®s grand nombre premier *(1)*
- $p-1$ a un tr√®s grand facteur premier *(2)*

Soit $q$ un nombre premier.

Supposons que $p=2q+1$ **et $p$ est un tr√®s grand nombre premier**.

Par construction, $p$ est un tr√®s grand nombre premier : la condition *(1)* est respect√©e.

$p=2q+1$

$p-1=2q$

Par hypoth√®se $q$ est un nombre premier et par construction $q$ est tr√®s grand.

Donc $p-1$ a un tr√®s grand facteur premier : la condition *(2)* est v√©rifi√©e.

Ainsi un tr√®s grand nombre premier de la forme $p=2q+1$ avec $q$ premier satisfait aux contraintes impos√©es √† $p$ dans ElGamal. 

```python
def PremierElGamal(k: int) -> int:
    """
    Retourne un nombre pseudo-premier p = 2q+1 o√π q est un nombre pseudo-premier cod√© sur k bits

    Args:
        k (int): Le nombre de bits sur lequel cod√© le facteur premier

    Returns:
        int: Un nombre pseudo-premier
    """

    epsilon = 1e-20

    q = Generateur_Premier(k, epsilon)
    p = 2 * q + 1

    # On fait passer p √† travers des tests de Miller afin de s'assurer qu'il est pseudo-premier aussi

    while not TestMiller(p, ceil(log((k * log(2)) / (2 * epsilon), 4))):
        # print(f'q: {q, TestMiller(q, ceil(log((k * log(2)) / (2 * epsilon), 4)))}, p: {p, TestMiller(p, ceil(log((k+1 * log(2)) / (2 * epsilon), 4)))}')
        q = Generateur_Premier(k, epsilon)
        p = 2 * q + 1

    return p
```

### 4.1.2 Construction d'une cl√© (p,a)

#### 4.1.2.1.

On a que $(p, a)$ est une cl√© ElGamal valide.

Donc $p$ est un tr√®s grand nombre premier.

Ainsi le groupe **multiplicatif** $\mathbb{Z}_p^* = \mathbb{Z}/p\mathbb{Z}$ est cyclique : fini et engendr√© par l'un de ses √©l√©ments $\xi \in \mathbb{Z}_p^*$ :

$\mathbb{Z}_p^* = \{1, \xi,..., \xi^{p-2}\}$

Or si $r$ est un √©l√©ment primitif modulo $p$, alors $r$ est un repr√©sentant de $\xi$ et v√©rifie $\xi = \bar{r}$.

Par hypoth√®se, $a$ est un √©l√©ment primitif modulo de p.

Donc $\bar{a} = \xi$ est un g√©n√©rateur du groupe multiplicatif $\mathbb{Z}_p^*$.

On explicite comme isomorphe entre le groupe **additif** $\mathbb{Z}_p^*$ et le groupe **multiplicatif** $\mathbb{Z}_p^*$, l'application exponentielle de base $a$ $exp_a$ d√©fini par :

$exp_a \colon x \in [ 1;p-1 ] \longmapsto a^x\ mod\ p \in [ 1;p-1 ]$

Par hypoth√®se, $p$ est premier et $a$ est un √©l√©ment primitif modulo $p$, donc l'application $exp_a$ admet une application r√©ciproque $log_a$ appel√©e logarithme discret de base $a$.

Comme $\bar{a} = \xi$ est un g√©n√©rateur du groupe multiplicatif $\mathbb{Z}_p^*$, $log_a(\bar{a})$ est un g√©n√©rateur du groupe **additif** $\mathbb{Z}_p^*$.

Par hypoth√®se, p est un tr√®s grand nombre premier et p ‚àí 1 a un tr√®s grand facteur premier.

Donc l'application $exp_a$ est √† sens unique: cet isomorphisme ne permet pas de construire $a$.

**En conclusion**, le groupe **multiplicatif** $\mathbb{Z}_p^*$ est isomorphe au groupe **additif** $\mathbb{Z}_p^*$ de g√©n√©rateur $log_a(\bar{a})$. Un isormorphisme qui convient est :

$exp_a \colon x \in [ 1;p-1 ] \longmapsto a^x\ mod\ p \in [ 1;p-1 ]$

$p$ est un tr√®s grand nombre premier.

Ainsi le groupe **multiplicatif** $\mathbb{Z}_p^* = \mathbb{Z}/p\mathbb{Z}$ est cyclique : fini et engendr√© par l'un de ses √©l√©ments $\xi \in \mathbb{Z}_p^*$ :

$\mathbb{Z}_p^* = \{1, \xi,..., \xi^{p-2}\}$.

Notons $n$ l'ordre de $\mathbb{Z}_p^*$ donc :

$n = p-1$

Or par le th√©or√®me de Lagrange (Th√©or√®me A.2.1 p.71) :
Pour tout $x \in \mathbb{Z}_p^*$, l'ordre de $x$ divise n.

Par construction :

$n = p-1$ or $p-1 = 2q$

$n = 2q$

Donc l'ordre de $x$ est dans l'ensemble $\{1; 2; q; 2q\}$.

```python
def CleElGamal(k: int) -> (int, int):
    """
    Retourne une cl√© de ElGamal (p, a)

    Args:
        k (int): Le nombre de bits sur lequel cod√© le facteur premier

    Returns:
        (int, int): Une cl√© de ElGamal (p, a)
    """

    # epsilon = 1e-20
    p = PremierElGamal(k)
    a = (p - 1) // 2

    return (p, a)
```

On a :
- $q$ est un tr√®s grand nombre premier
- $q \in \mathbb{Z}_p^*$

Donc choisir $a = q$ est valable. C'est-√†-dire $a = \frac{p-1}{2}$.

```python
def ChiffrementElGamal(
    message: str, cleCommune: (int, int), clePubliqueDestinataire: int, clePrivee: int
) -> int:
    """
    Chiffre par ElGamal un message clair en fournissant la cl√© publique commune, la cl√© publique du destinataire, la cl√© priv√©e de l'√©metteur

    Args:
        message (str): Le message en clair
        cleCommune (int, int): Une cl√© commune d'ElGamal (p,a)
        clePubliqueDestinataire (int): La cl√© publique du destinaire
        clePrivee (int): La cl√© priv√©e de l'√©metteur

    Returns:
        int: Les codes du chiffr√© par ElGamal
    """

    p, a = cleCommune
    k = ExpoModRapide(clePubliqueDestinataire, clePrivee, p)

    codesClairs = Encodage(message, 3)

    codesChiffres = [k * code for code in codesClairs]

    return codesChiffres
```

```python
def DechiffrementElGamal(
    codesChiffres: [int],
    cleCommune: (int, int),
    clePubliqueEmetteur: int,
    clePrivee: int,
) -> str:
    """
    D√©hiffre par ElGamal un message cod√© en fournissant la cl√© publique commune, la cl√© publique de l'√©m√©tteur, la cl√© priv√©e du destinataire

    Args:
        codesChiffres (int]): La liste des codes des messages chiffr√©s par ElGamal
        cleCommune (int, int): Une cl√© commune d'ElGamal (p,a)
        clePubliqueEmetteur (int): La cl√© publique de l'√©m√©tteur
        clePrivee (int): La cl√© priv√©e du destinataire

    Returns:
        str: Le message d√©chiffr√© par ElGamal
    """

    p, a = cleCommune
    k = ExpoModRapide(clePubliqueEmetteur, clePrivee, p)
    kInverse = InverseModulo(k, p)

    codesClairs = [(kInverse * code) % p for code in codesChiffres]

    messageDecode = Decodage(codesClairs)

    return messageDecode
```

```python
texteClair = "- Je suis Alexandre\n- Et moi Diog√®ne, le cynique\n- Demande-moi ce que tu veux, je te le donnerai\n- Alors : ote-toi de mon soleil"
print(f"Alice va envoyer le message suivant:\n{texteClair}")

chiffrageElGamal = ChiffrementElGamal(texteClair, (p, a), k_B, A)
dechiffrageElGamal = DechiffrementElGamal(chiffrageElGamal, (p, a), k_A, B)

print(
    f"Apr√®s chiffrage par Alice et d√©chiffrage par Bob, ce dernier a re√ßu le message suivant:\n{dechiffrageElGamal}"
)
print(
    f"Les messages sont donc{'' if texteClair==dechiffrageElGamal else 'non'} identiques."
)
```

```python
F2 = IntegerModRing(2)
X = polygen(F2, "X")

mPolynome = X**8 + X**4 + X**3 + X + 1


def Car2Bin(char: str) -> str:
    char = char[0]
    charUnicode = ord(char)
    charBinaire = format(charUnicode, "b").zfill(8)

    return charBinaire


def Bin2PolF2(charBinaire: str):
    polynome = 0

    for i in range(
        8
    ):  # Dans l'√©ventualit√© o√π le codage en bits d√©passe 8 bits, on tronque
        bit = charBinaire[::-1][
            i
        ]  # On inverse le chaine de bits pour pouvoir lire du bit de poids faible jusqu'au bit de poids fort
        if bit == "1":
            polynome = polynome + X**i

    return polynome


def PolF22Bin(polynome) -> str:
    listeBits = list(polynome)

    charBinaire = ""

    for bit in listeBits:
        charBinaire = str(bit) + charBinaire

    charBinaire = charBinaire

    return charBinaire


def Car2PolF2(char: str):
    charBinaire = Car2Bin(char)
    polynome = Bin2PolF2(charBinaire)

    return polynome
```

```python
def DecoupeBlocs(message: str) -> [str]:
    nombreBlocs = len(message) // 16

    blocs = []

    for i in range(nombreBlocs):
        bloc = message[i * 16 : (i + 1) * 16]
        blocs.append(bloc)

    longueurDernierblocModulo16 = len(message) % 16
    """
    Si la longueur du message n'est pas un multiple de 16 alors le dernier bloc n'a pas √©t√© trait√©, et il faut faire une op√©ration de rembourrage.
    """
    if (longueurDernierblocModulo16) != 0:
        dernierBloc = message[
            nombreBlocs * 16 :
        ]  # R√©cup√©ration du dernier bloc incomplet
        dernierBloc = dernierBloc + "\0" * (
            16 - longueurDernierblocModulo16
        )  # Rembourrage pour obtenir un bloc de 16 caract√®res
        blocs.append(dernierBloc)

    return blocs
```

```python
blocs = DecoupeBlocs(texteClair)
print(blocs)
```

```python
def Bloc2Bin(bloc: str) -> str:
    blocBinaire = ""

    for char in bloc:
        charUnicode = ord(char)
        charBinaire = format(charUnicode, "b").zfill(8)
        charBinaire = charBinaire[
            :8
        ]  # On tronque dans l'√©ventualit√© o√π la chaine de caract√®re √©tait cod√© sur plus de 8 bits

        blocBinaire = (
            blocBinaire + charBinaire
        )  # On ajoute l'octet en respectant l'ordre des caract√®res dans le bloc

    return blocBinaire


def BlocsEnBinaire(blocs: [str]) -> [str]:
    blocsBinaire = [Bloc2Bin(bloc) for bloc in blocs]

    return blocsBinaire
```
